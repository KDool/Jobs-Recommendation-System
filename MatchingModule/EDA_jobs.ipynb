```python
# Filter a string Jobtile include a role
def filter_job_role(jobTitle='',role= ''):
    if role.lower() in jobTitle.lower():
        return 1
    else:
        return 0

# Filter DataFrame of Jobs with Role Input

def filter_DataFrame_role(df:pd.DataFrame,input_role=''):
    df_x = df.copy()
    column_check = []
    for i in range(0,len(df_x)):
        var_check = filter_job_role(df_x.iloc[i]['jobTitle'],input_role)
        column_check.append(var_check)
    df_x['column_check'] = column_check
    df_y = df_x[df_x['column_check']==1]
    df_y = df_y[['_id','jobTitle','location','company','type','desc']]
    return df_y

df_DE = filter_DataFrame_role(df_jobs,'engineer')
df_DE
```

```python
import collections
def filter_location(text=''):
    l1 = list(set(text.split(',')))
    r = []
    for i in range(0,len(l1)):
        l1[i] = l1[i].strip().lower()
    l1 = list(set(l1))
    for i in range(0,len(l1)):
        if '/' not in l1[i]:
            r.append(l1[i])
    # print("L1: ",r)
    return r

# filter_location('Hanoi Capital Region')

def EDA_on_location(df:pd.DataFrame):
    df_x = df.copy()
    temp_location = []
    for i in range(0,len(df_x)):
        text = df.iloc[i]['location']
        temp_location.append(filter_location(text))
    df_x['filter_location'] = temp_location
    list_of_types = []

    for i in range (0, len(df_x)):
        list_of_types += df_x.iloc[i]['filter_location']
    dict_result = dict(collections.Counter(list_of_types))
    # print(dict_result)
    keys = list(dict_result.keys())
    values = list(dict_result.values())
    return keys,values
    
```

```python
def filter_workingType(text=''):
    l1 = list(set(text.split('·')))
    r = []
    for i in range(0,len(l1)):
        l1[i] = l1[i].strip().lower()
    l1 = list(set(l1))
    for i in range(0,len(l1)):
        if '/' not in l1[i]:
            r.append(l1[i])
    # print("L1: ",r)
    return r

filter_workingType('Internship · Internship')
```

```python
def process_DfWorkingType(df:pd.DataFrame):
    column_types = []
    df_x = df.copy()
    for i in range(0, len(df)):
        column_types.append(filter_workingType(df_x.iloc[i]['type']))
    df_x['working'] = column_types
    
    return df_x

df_working = process_DfWorkingType(df_jobs)
```
