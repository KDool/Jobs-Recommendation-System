```python
import pandasql as ps
import duckdb

df_output = duckdb.query("SELECT df1._id, df1.skills,df2.education_skills FROM df1,df2 WHERE df1._id = df2._id").df()
# output = sqldf("select _id from df_edu2skills", locals())
```

```python
print(type(df_output))
df_output
```

```python
import ast
for i in range(0,len(df_output)):
    # df_output.iloc[i]['education_skills'] = ast.literal_eval(df_output.iloc[i]['education_skills'])
    df_output.iloc[i]['skills'] = ast.literal_eval(df_output.iloc[i]['skills'])
    df_output.iloc[i]['education_skills'] = ast.literal_eval(df_output.iloc[i]['education_skills'])
```

```python
for i in range(0,len(df_output)):
    user_skills = df_output.iloc[i]['skills']
    # print(user_skills)
    # for item in user_skills:
    #     item = item.lower()
    for j in range(len(user_skills)):
        user_skills[j] = user_skills[j].lower()
    df_output.iloc[i]['skills'] = user_skills
```

```python
df_output
```

```python
user_vector_list = []
for i in range(0,len(df_output)):
    user_vector = []
    user_skills = df_output.iloc[i]['skills']
    user_education_skills = df_output.iloc[i]['education_skills']
    user_vector = user_skills + user_education_skills
    user_vector = list(set(user_vector))
    user_vector_list.append(user_vector)
df_output['user_vector'] = user_vector_list
```
